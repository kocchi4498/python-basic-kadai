```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 1〜100のランダムな整数を作る
var = random.randint(1, 100)

# 変数varの値を出力
print(var)

#変数varが、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力 変数varが、3の倍数の場合は「Fizz」を出力 変数varが、5の倍数の場合は「Buzz」を出力
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
elif var % 3 == 0:
    print("Fizz")
elif var % 5 == 0:
    print("Buzz")
else:
    print(var)
```

```python
# 消費税を計算する関数
def calc_tax(price, tax):
    total = price + price * tax / 100
    return total

# 関数を呼び出す
result = calc_tax(110, 10)

# 結果を表示
print(result)
```
